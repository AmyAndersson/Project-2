```python
species_DB2 = species_DB[species_DB["Species"] != "Na"]
species_DB3 = species_DB2[species_DB2["Common Name"] != "Na"]
species_DB4 = species_DB3[species_DB3["Category"] != "Na"]
species_DB5 = species_DB4[species_DB4["Kingdom"] != "Na"]
species_DB6 = species_DB5[species_DB5["Class"] != "Na"]
species_DB7 = species_DB6[species_DB6["Family"] != "Na"]
species_DB8 = species_DB7[species_DB7["Order"] != "Na"]
species_DB9 = species_DB8[species_DB8["Phylum"] != "Na"]
species_DB10 = species_DB9[species_DB9["Type of Presence"] != "Na"]
species_DB10
```

```python
#'Species', 'Common Name', 'Category', 'Type of Presence',
#'Location(s)', 'Kingdom', 'Phylum', 'Class', 'Order', 'Family',


species_DB10["Category"].unique()
```

```python
species_DB10 = species_DB10.rename(columns = {'Type of Presence': 'Presence_Type','Common Name': 'Common_Name', 'Location(s)': 'Location'}, inplace = False)
```

```python
species_DB10
```

```python
species_cleaned = species_DB10.loc[:, ["Species", "Common_Name", "Category",
                                    "Presence_Type", "Location",  'Kingdom', "Phylum", 'Class', "Order", 
       'Family']]
```
