```python
#bring in all the libraries we need to be able to connect to the database so we can export the data to SQL
from sqlalchemy import create_engine
from config import username
from config import password
from config import database
import pandas as pd

#create engine to talk to the database
engine = create_engine(f"postgresql://{username}:{password}@localhost:5432/{database}")
conn=engine.connect()
```

```python
file_to_load = "report.csv"
```

```python
#show what data we have to work with, we need to skip the first header line as this has no data of value
enviro_data = pd.read_csv(file_to_load,skiprows=1)
enviro_data.head()
```

```python
#drop any fields with no data
dropnan = enviro_data.dropna(axis=1, how="all")
dropnan
```

```python
#create a new df with only the columns that we might use for visualisations
datd_df = dropnan[["Scientific Name","Common Name","EPBC Threat Status","EPBC Threatened Species Listed Name","EPBC Threatened Species Date Effective","Kingdom","Phylum","Class","Order","Family","Genus"]]
datd_df
```

```python
#we want to focus mainly on the data that shows the threat status so remove any rows that have no data
clean_data = datd_df[datd_df["EPBC Threat Status"].notnull()]
clean_data
```
